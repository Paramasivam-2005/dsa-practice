**994.Rotten Orenges**

```python
# using BFS to solve the problem
# Time Complexity: O(m*n)
# Space Complexity: O(m*n)
# where m is number of rows and n is number of columns in the grid
# we use a queue to keep track of the rotten oranges and for each rotten orange we check its 4 adjacent cells
from typing import List
from collections import deque
class Solution:
    def orangesRotting(self, grid: List[List[int]]) -> int:
        row=len(grid)
        col=len(grid[0])
        min,fresh=0,0
        direction=[(-1,0),(1,0),(0,-1),(0,1)]
        q=deque()
        for r in range(row):
            for c in range(col):
                if grid[r][c]==2:
                    q.append((r,c))
                elif grid[r][c]==1:
                    fresh+=1    
        while q and fresh>0:
            for _ in range(len(q)):
                r,c=q.popleft()
                for dr,dc in direction:
                    nr,nc=r+dr,c+dc
                    if(nr>=0 and nr<row and nc>=0 and nc<col and grid[nr][nc]==1):
                        q.append((nr,nc))
                        fresh-=1
                        grid[nr][nc]=2
            min+=1
        print(min)    
        return min if fresh==0 else -1    
```

**1020. Number of Enclaves**

```python
from typing import List

class Solution:
    def numEnclaves(self, grid: List[List[int]]) -> int:

        def dfs(r, c):
            if r < 0 or r >= row or c < 0 or c >= col or grid[r][c] != 1:
                return
            
            grid[r][c] = 0   # mark as visited / water

            dfs(r-1, c)
            dfs(r+1, c)
            dfs(r, c-1)
            dfs(r, c+1)

        row, col = len(grid), len(grid[0])

        # 1️⃣ Top boundary
        for c in range(col):
            if grid[0][c] == 1:
                dfs(0, c)

        # 2️⃣ Bottom boundary
        for c in range(col):
            if grid[row-1][c] == 1:
                dfs(row-1, c)

        # 3️⃣ Left boundary
        for r in range(row):
            if grid[r][0] == 1:
                dfs(r, 0)

        # 4️⃣ Right boundary
        for r in range(row):
            if grid[r][col-1] == 1:
                dfs(r, col-1)

        # 5️⃣ Count remaining land
        count = 0
        for r in range(row):
            for c in range(col):
                if grid[r][c] == 1:
                    count += 1

        return count
```
