***Permutation***
- **C(n,r)=C(n−1,r−1)+C(n−1,r)**

```python
#include <stdio.h>
int per(int n,int r){
    if(r==0||r==n) return 1;
    if (r>n) return 0;
    return per(n-1,r-1)+per(n-1,r);
}
int main(){
    int n;
    int r;
    scanf("%d",&n);
    scanf("%d",&r);
    printf("%d",per(n,r));
    
}
```

***Circular Prime***
- n=197
- last=n%10  Then last=7
- n= n/10 + 10^3*last ===> 19+700=719 

```python
#include <stdio.h>
#include <math.h>
int prime(int n){
    int flag=1;
    for(int i=2;i*i<=n;i++){
        if(n%i==0) flag=0;
    }
    return flag;
}
int main(){
    int n;
    scanf("%d",&n);
    int l=log10(n)+1;
    int pow10=pow(10,l-1);
    int temp=n;
    for(int i=0;i<l;i++){
        if(!prime(temp)){
            printf("not circular prime");
            return 0;
        }
        int last=temp%10;
        temp=(temp/10) + (last*pow10);
    }
    printf("jaajjds");
}
```
