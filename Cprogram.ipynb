***Permutation***
- **C(n,r)=C(n−1,r−1)+C(n−1,r)**

```python
#include <stdio.h>
int per(int n,int r){
    if(r==0||r==n) return 1;
    if (r>n) return 0;
    return per(n-1,r-1)+per(n-1,r);
}
int main(){
    int n;
    int r;
    scanf("%d",&n);
    scanf("%d",&r);
    printf("%d",per(n,r));
    
}
```
